```python
import spotipy
from spotipy.oauth2 import SpotifyClientCredentials
import pandas as pd
from pandas import json_normalize
from random import randint
from time import sleep
```

```python
secrets_file = open(".gitignore/secrets.txt","r")
string = secrets_file.read()
secrets_dict={}
for line in string.split('\n'):
    if len(line) > 0:
        #print(line.split(':'))
        secrets_dict[line.split(':')[0]]=line.split(':')[1].strip()
sp = spotipy.Spotify(auth_manager=SpotifyClientCredentials(client_id=secrets_dict['clientid'],
                                                           client_secret=secrets_dict['clientsecret']))
```

```python
def get_playlist_tracks(playlist_id):
    results = sp.user_playlist_tracks("spotify",playlist_id)
    tracks = results['items']
    while results['next']!=None:
        results = sp.next(results)
        tracks = tracks + results['items']
    return tracks
```

```python
all_tracks = get_playlist_tracks("1dhJtgwleWL9LToqQTqddH")
len(all_tracks)
```

```python
tracks2 = json_normalize(all_tracks)
tracks2
```

```python
nan_values = tracks2['track.artists'].isna()
print(tracks2[nan_values])
```

```python
rows_to_drop = [29, 1731, 3445, 5159, 6873]
tracks2 = tracks2.drop(rows_to_drop)
```

```python
print(tracks2[nan_values])
```

```python
tracks2.reset_index(drop=True, inplace=True)
```

```python
artists_df = pd.DataFrame(columns=['href', 'id', 'name', 'type', 'uri', 'external_urls.spotify','song_id', 'song_name', 'popularity' ])
for i in tracks2.index:
    artists_for_song = json_normalize(tracks2.iloc[i]['track.artists'])
    artists_for_song['song_id']    = tracks2.iloc[i]['track.id'] 
    artists_for_song['song_name']  = tracks2.iloc[i]['track.name']  
    artists_for_song['popularity'] = tracks2.iloc[i]['track.popularity'] 
    artists_df = pd.concat([artists_df, artists_for_song], axis=0)
```

```python
artists_df
```

```python
df_final = artists_df[['song_name', 'name', 'song_id', 'popularity']]
df_final
```

```python
chunks = [(i, i+100) for i in range(0, len(df_final), 100)]
chunks
audio_features_list = []
for chunk in chunks:
    id_list100 = df_final['song_id'][chunk[0]:chunk[1]]
    audio_features_list = audio_features_list + sp.audio_features(id_list100)
    sleep(randint(1,3000)/1000)
len(audio_features_list)
```

```python
audio_features_df = json_normalize(audio_features_list)
```

```python
audio_features_df.drop_duplicates(inplace=True)
```

```python
df_w_audio_ft = pd.merge(left=df_final,
                        right=audio_features_df,
                        how='inner',
                        left_on='song_id',
                        right_on='id')
df_w_audio_ft
```

```python
df_w_audio_ft.to_csv('music.csv', index=False)
```
